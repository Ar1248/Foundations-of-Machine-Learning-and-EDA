 1. What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each. (Hint: Compare their scope, techniques, and applications for each.)

  - Artificial Intelligence (AI), Machine Learning (ML), Deep Learning (DL), and Data Science are connected fields but differ in scope, techniques, and applications.
  
  - They can be visualized hierarchically :
    - Data Science is the broadest domain involving data analysis
      - AI covers intelligent system creation
        - ML and DL are core techniques within AI.

  - Artificial Intelligence (AI)
    - AI is the broad field concerned with creating machines that can perform tasks requiring human-like intelligence, such as reasoning, perception, and decision-making.​

    - Scope: Broad—includes logic-based systems, rule-based automation, robotics, and learning systems.

    - Techniques: Knowledge graphs, expert systems, natural language processing, reinforcement learning, and generative models.

    - Applications: Virtual assistants (e.g., Alexa, Siri), autonomous vehicles, fraud detection, and healthcare diagnostics.

  - Machine Learning (ML)
    - ML is a subset of AI focused on developing algorithms that enable computers to learn from experience (data) and improve performance without explicit programming.​

    - Scope: Narrower than AI—focuses purely on pattern recognition and predictive analytics.

    - Techniques: Supervised, unsupervised, and reinforcement learning methods such as regression, classification, and clustering.

    - Applications: Recommendation engines (Netflix, Amazon), spam detection, predictive maintenance, and fraud detection.

  - Deep Learning (DL)
    - DL is a subset of ML that uses multi-layered neural networks to automatically extract complex patterns and features from large amounts of data, mimicking human learning.​

    - Scope: Focused primarily on unstructured, high-dimensional data.

    - Techniques: Convolutional Neural Networks (CNNs), Recurrent Neural Networks (RNNs), and Generative Adversarial Networks (GANs).

    - Applications: Image and speech recognition, natural language translation, and content generation.

  - Data Science
    - Data Science is broader than all three—its main goal is to extract actionable insights from data using statistical analysis, data mining, and modeling.​

    - Scope: Covers the entire data lifecycle—collection, preprocessing, analysis, and visualization.

    - Techniques: Statistical modeling, hypothesis testing, machine learning integration, and data visualization tools.

    - Applications: Business forecasting, customer segmentation, healthcare analytics, and scientific research.

2. Explain overfitting and underfitting in ML. How can you detect and prevent
them? ( Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques. )

  - Overfitting
    - Overfitting occurs when a model learns the training data too well, including its noise and outliers, which makes it perform excellently on training data but poorly on new, unseen data. This usually happens with highly complex models that have high variance and low bias.

      - Detection: High accuracy on training data but low accuracy on validation/test data.

      - Prevention: Use regularization techniques like L1 (Lasso) or L2 (Ridge) to penalize large coefficients, reduce model complexity, and apply dropout in neural networks. Increasing training data size also helps. Cross-validation (e.g., k-fold) estimates model generalization better and mitigates overfitting by validating on several data subsets.

  - Underfitting
    - Underfitting occurs when the model is too simple to capture the underlying patterns in the data, resulting in poor performance on both training and test data. This happens with models having high bias and low variance.

      - Detection: Poor accuracy on both training and validation/test data.

      - Prevention: Increase model complexity, add relevant features, decrease regularization, or use more sophisticated algorithms.

  - Bias-Variance Tradeoff
    - The bias-variance tradeoff is the balance between simplicity and complexity of a model.

      - Models with high bias (simpler models) tend to underfit (miss relevant relations).

      - Models with high variance (more complex models) tend to overfit (capture noise). The ideal model skillfully balances bias and variance to minimize total error.
  - Cross-validation helps estimate a model’s ability to generalize by training and testing on different subsets of data, ensuring better balance.
  - Regularization techniques add penalties to avoid overly complex models, helping to reduce overfitting without a substantial increase in bias.

3. :How would you handle missing values in a dataset? Explain at least three
methods with examples.
( Hint: Consider deletion, mean/median imputation, and predictive modeling. )
  - There are multiple ways to handle missing data, depending on how much data is missing and its importance. Here are three commonly used methods with examples:
    1. Deletion (Listwise or Pairwise)
      - Description: This method removes rows or columns that contain missing values when the proportion of missing data is small and occurs randomly.​

        - Row-wise deletion: Deletes entire records (rows) with any missing values.

        - Column-wise deletion: Removes columns with too many missing entries.

    2. Imputation (Mean/Median/Mode)
      - Description: Replace missing values with estimated values computed from the non-missing data.

        - Mean or median imputation: Used for numerical data. The mean is used if data distribution is symmetric, and the median is used if skewed.​

        - Mode imputation: Used for categorical data, replacing missing values with the most frequent category.
    3. Predictive Modeling (Machine Learning-Based Imputation)
      - Description: Uses models like regression, K-Nearest Neighbors (KNN), Decision Trees, or Random Forests to predict missing values based on other features.​

        - Linear regression imputation: Predicts missing numerical values using correlated features.

        - KNN imputation: Fills in missing values based on average values from nearest data points.

        - Random Forest or MICE (Multiple Imputation by Chained Equations): Uses iterative models for complex patterns and correlated features.

4. What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).( Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.)
  - An imbalanced dataset in machine learning refers to a dataset where the distribution of classes is uneven, with one class (the majority class) having significantly more samples than another (the minority class). This imbalance can lead to biased models that predict the majority class well but perform poorly on the minority class, which is often the class of interest, such as fraud detection or disease diagnosis.​

  - Two Techniques to Handle Imbalanced Datasets
  
    - SMOTE (Synthetic Minority Over-sampling Technique)

      - Theoretical: SMOTE generates synthetic examples for the minority class by interpolating between existing minority samples to balance the class distribution. This avoids simple duplication of minority samples and helps the model learn better decision boundaries.

      - Practical: Using Python libraries like imbalanced-learn, SMOTE can be applied to create balanced training sets, improving model sensitivity to the minority class.

    - Random Under/Oversampling

      - Random Oversampling: This method duplicates examples from the minority class to increase its representation.

      - Random Undersampling: This reduces the majority class by randomly removing samples to balance the dataset.

      - Practical: Oversampling is useful when data is scarce but can lead to overfitting. Undersampling reduces training time but risks losing important information. These methods are easily implemented using tools like scikit-learn or imbalanced-learn.

    - Class Weights in Models

        - Theoretical: Models are modified to assign higher penalty (weight) to misclassifications of the minority class during training. This forces the algorithm to prioritize learning the minority class.

        - Practical: Many ML libraries (e.g., scikit-learn, XGBoost) support class weights or cost-sensitive learning, improving performance on imbalanced data without altering the data distribution.

5.  Why is feature scaling important in ML? Compare Min-Max scaling and Standardization. (Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.)
  - Feature scaling is a preprocessing technique that transforms numerical features so they exist on a similar range or unit, ensuring all features contribute equally to a machine learning model’s performance. It is particularly crucial for algorithms that rely on distances or gradients.
  Thus, Different features in raw data can have vastly different scales — for example, age (0–100) and income (0–100,000). Without scaling:

    - Distance-based algorithms like K-Nearest Neighbors (KNN) or Support Vector Machines (SVM) may become biased toward features with larger magnitudes because they dominate the distance metric.

    - Gradient descent–based algorithms (used in Linear/Logistic Regression or Neural Networks) may converge slowly or inefficiently, as one feature’s steeper gradient can overshadow others.​

    - Regularization penalties (L1/L2) can also be misapplied if scales differ.

    Feature scaling thus improves convergence, prevents feature dominance, and enhances accuracy in many ML models.

  - Comparison :

    - Min-Max Scaling (Normalization)
      - Formula : x′ = {x - x(min)} / {x(max)-x(min)}
      - Output Range : rapidops (sometimes [-1, 1])
      - When to Use	: When the algorithm requires bounded features (e.g., Neural Networks, KNN, SVM)
      - Effect on Outliers : Sensitive — outliers can distort min/max values
      - Example (Python)	: Using MinMaxScaler() from scikit-learn
      - Ideal For :	Image processing, deep learning models
    
    - Standardization (Z-score Normalization)
      - Formula : x' = (x - µ )/ σ
      - Output Range : Mean = 0, Standard deviation = 1
      - When to Use	: When the data follows a Gaussian distribution or algorithms assume centered data
      - Effect on Outliers : More robust — large values don't affect scaling as much
      - Example (Python)	: Using StandardScaler()  from scikit-learn
      - Ideal For :	PCA, Logistic/Linear Regression, SVMs


6. Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.

  - Label Encoding and One-Hot Encoding are both techniques to convert categorical variables into numerical data, but they differ fundamentally in use cases and representation.

  - Label Encoding
    - Assigns each category a unique integer value in a single feature/column.

    - Best suited for ordinal categorical variables where categories have an inherent order (e.g., "Low", "Medium", "High").

    - Keeps dimensionality low by creating just one encoded column.

    - Works well with tree-based models (like decision trees, random forests) that can handle ordinal features.

    - However, it can mislead models that interpret numeric values as continuous or ordered if categories are nominal.

  - One-Hot Encoding
    - Creates multiple binary columns, one per category, with 1 indicating presence and 0 absence.

    - Best suited for nominal categorical variables that have no natural order (e.g., "red", "blue", "green").

    - Increases dimensionality, which may add computational cost and sparsity.

    - Preferred for algorithms that do not assume ordinality, such as logistic regression, neural networks, and KNN.

    - Prevents models from inferring false order or priority among categories.

9. Flight Price Prediction Dataset

a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window.

b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why?

Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)

SOLUTION :
  - Flight Prices vs Days Left Until Departure
    - Flight prices tend to be higher very close to the departure date (around 1-2 days left), with prices averaging above ₹21,000 to ₹30,000 in the initial days.

    - Prices fluctuate but generally decrease to an average around ₹18,500 to ₹19,500 as the booking window extends towards 45-49 days before departure.

    - This pattern indicates price surges near departure dates, likely due to increased demand and reduced seat availability.

    - Recommended Booking Window: Booking flights closer to 1-2 months ahead tends to offer lower prices, avoiding last-minute exponential price surges.

  - Price Comparison Across Airlines for Delhi-Mumbai Route
    - Airlines like AirAsia, Indigo, and SpiceJet are consistently cheaper and positioned as low-cost carriers.

    - Airlines like Air India and Vistara have much higher average prices, reflecting their premium service positioning, broader amenities, and possibly legacy carrier status.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
flight_data = pd.read_csv('flight_price.csv')

# a) Price variation with days left to departure
days_price_avg = flight_data.groupby('days_left')['price'].mean().reset_index()

# b) Compare prices for Delhi-Mumbai across airlines
route_flights = flight_data[(flight_data['source_city'] == 'Delhi') & (flight_data['destination_city'] == 'Mumbai')]
airline_price_avg = route_flights.groupby('airline')['price'].mean().sort_values()

# Visualization (use seaborn and matplotlib in local environment)
sns.lineplot(data=days_price_avg, x='days_left', y='price', color='green')
plt.title('Average Flight Price vs Days Left Until Departure')
plt.xlabel('Days Left Until Departure')
plt.ylabel('Average Price')
plt.show()


# Use a color palette
colors = sns.color_palette("Set2", n_colors=len(airline_price_avg))

# Create a bar plot with colorful bars
sns.barplot(x=airline_price_avg.index, y=airline_price_avg.values, palette=colors)

# Add labels to bars
for index, value in enumerate(airline_price_avg):
    plt.text(index, value + 300, f'₹{int(value)}', ha='center', va='bottom')

plt.title('Average Flight Price by Airline (Delhi to Mumbai)')
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

a). What factors most strongly correlate with employee attrition? Use visualizations to show keydrivers (e.g., satisfaction, overtime, salary).

b). Are employees with more projects more likely to leave?

Dataset: hr_analytics

SOLUTION :

  - Factors Correlated with Employee Attrition
    - Key Insight:
      - The strongest driver of attrition is low satisfaction level, followed by long tenure and excessive working hours, which often indicate overwork or lack of recognition.

  - Projects and Attrition Relationship

      - Employees with too few (2) or too many (6-7) projects show much higher attrition rates.

      - The ideal workload range (3–4 projects) corresponds to the lowest attrition rates, suggesting balanced workloads help retain employees.

      - Extremely high attrition for employees managing 6+ projects indicates burnout.
  - Recommendations:

    - Enhance employee satisfaction through better compensation and recognition.

    - Moderate project assignments to the 3–4 range for optimal engagement.

    - Monitor overworked or underutilized employees for early intervention [attached_file:hr_analytics.csv].

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

hr = pd.read_csv('hr_analytics.csv')

# Correlation with attrition
correlations = hr.corr(numeric_only=True)['left'].sort_values(ascending=False)
sns.barplot(x=correlations.index, y=correlations.values, palette='coolwarm')
plt.title('Correlation of Features with Employee Attrition')
plt.ylabel('Correlation with Attrition')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Attrition vs number of projects
project_attrition = hr.groupby('number_project')['left'].mean().reset_index()
sns.barplot(data=project_attrition, x='number_project', y='left', palette='Set2')
plt.title('Attrition Rate vs Number of Projects')
plt.xlabel('Number of Projects')
plt.ylabel('Attrition Rate')
plt.tight_layout()
plt.show()
```
